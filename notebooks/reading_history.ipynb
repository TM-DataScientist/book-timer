# Reading History Viewer

`data/reading_history.parquet` を読み込んで、読了履歴を確認するための Notebook です。

```python
from pathlib import Path

import pandas as pd

history_path = Path("..") / "data" / "reading_history.parquet"

if history_path.exists():
    history_df = pd.read_parquet(history_path)
    history_df = history_df.sort_values(
        by=["session_date", "book_title"],
        ascending=[False, True],
    ).reset_index(drop=True)
    history_df
else:
    print(f"読了履歴ファイルが見つかりません: {history_path}")
```
